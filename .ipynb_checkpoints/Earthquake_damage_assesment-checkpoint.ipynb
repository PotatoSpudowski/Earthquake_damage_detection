```python
def IBM_Vision(image):
    url = 'https://gateway.watsonplatform.net/visual-recognition/api/v3/classify?version=2018-03-19'
    files = {'images_file': open(image,'rb')}
    resp = requests.post(url, auth=('apikey', '************'), files=files)
    count = 0
    results = resp.json()
    results = results['images'][0]['classifiers'][0]['classes']
    for result in results:
        if (result['class'] == "Earthquake"):
            return "Damage detected"
        
            
```

After using IBM's Api we also use our own algorithm to compare 2 satellite images to detect changes. If the change is too high ie there has been a major 

```python
def Eyeframe_Imchange(im1,im2):
    
    def DistMap(frame1, frame2):
        frame1_32 = np.float32(frame1)
        frame2_32 = np.float32(frame2)
        diff32 = frame1_32 - frame2_32
        norm32 = np.sqrt(diff32[:,:,0]**2 + diff32[:,:,1]**2 + diff32[:,:,2]**2)/np.sqrt(255**2 + 255**2 + 255**2)
        dist = np.uint8(norm32*255)

        return dist

    dist = DistMap(im1, im2)
    mod = cv2.GaussianBlur(dist, (9,9), 0)
    _, thresh = cv2.threshold(mod, 100, 255, 0)
    _, stDev = cv2.meanStdDev(mod)
    plt.imshow(mod)
    plt.title('Change')
    plt.show()
    return stDev
```

```python
image1 = cv2.imread('before.jpg')
image2 = cv2.imread('After.jpg')
print(Eyeframe_Imchange(image1,image2))
```

```python
sdThresh = 18
val = Eyeframe_Imchange(image1,image2)
if val > sdThresh:
    print("Earthquake damage detected")
```
